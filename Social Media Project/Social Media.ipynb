```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
import numpy as np
```

Load Data

```python
table=pd.read_csv("E:/Python Files/Social Media Marketing/SocialMediaData.csv", encoding="latin1")
```

Data Processing

```python
table.isnull().sum
table=table.dropna()
print(table.head())

impressions=table["Impressions"].sum()
home = table["From Home"].sum()
hashtags = table["From Hashtags"].sum()
Search = table["From Search"].sum()
other = table["From Other"].sum()

labels = ['From Home','From Hashtags','From Search','Other']
values = [home, hashtags, Search, other]
product_i=table.groupby('Product')['Impressions'].sum().reset_index()
```

Product Performance In Social Media

```python
plt.figure(figsize=(10, 10))
plt.pie(
    values,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90
    )

plt.title('Product performance in social media')

plt.axis('equal')

plt.show()
```

Impressions on Instagram Posts From Various Sources

```python
plt.figure(figsize=(10,10))

plt.pie(
    values if not isinstance(values, str) else table[values],
    labels=labels if not isinstance(labels, str) else table[labels],
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Impressions on Instagram Posts From Various Sources')
plt.axis('equal')

plt.show()
```

Relationship Between Likes and Impressions

```python
x = table["Impressions"]
y = table["Likes"]

plt.figure(figsize=(18, 6))

sizes = y * 2

scatter = plt.scatter(
    x,
    y,
    s=sizes,
    c=y,
    cmap='viridis',
    alpha=0.6,
    edgecolors='black',
    linewidths=0.5
    )

plt.colorbar(scatter, label="Likes")

m, b = np.polyfit(x, y, 1)
sorted_idx = np.argsort(x)
x_sorted = x.iloc[sorted_idx]
y_pred = m * x_sorted + b

plt.plot(x_sorted, y_pred)

plt.xlabel("Impressions")
plt.ylabel("Likes")
plt.title("Relationship Between Likes and Impressions")

plt.show()
```

Shares vs impressions

```python
x = table["Impressions"]
y = table["Shares"]

plt.figure(figsize=(18, 6))

sizes = y * 5

scatter = plt.scatter(
    x,
    y,
    s=sizes,
    c=y,
    cmap='viridis',
    alpha=0.6
    )

plt.colorbar(scatter, label="Shares")

m, b = np.polyfit(x, y, 1)
x_sorted = np.sort(x)
plt.plot(x_sorted, m*x_sorted + b)

plt.xlabel("Impressions")
plt.ylabel("Shares")
plt.title("Shares vs impressions")

plt.show()
```

Profile Views vs Follower Gains

```python
x = table["Profile Visits"]
y = table["Follows"]

plt.figure(figsize=(18, 6))

sizes = y * 5

scatter = plt.scatter(
    x,
    y,
    s=sizes,
    c=y,
    cmap='viridis',
    alpha=0.6,
    edgecolors='black',
    linewidths=0.5
)

plt.colorbar(scatter, label="Follows")

m, b = np.polyfit(x, y, 1)
sorted_idx = np.argsort(x)
x_sorted = x.iloc[sorted_idx]
y_pred = m * x_sorted + b

plt.plot(x_sorted, y_pred)

plt.xlabel("Profile Visits")
plt.ylabel("Follows")
plt.title("Profile Views vs Follower Gains")

plt.show()
```
